```python
import time
import jieba
import string
import gensim
from joblib import dump
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from zhon.hanzi import punctuation
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.decomposition import PCA
```

```python
## config
DATA_PATH = '../data/data.csv'
CONTENT_FIELD = 'content'
# LABEL_FIELD = 'class_label'          
STOPWORDS_PATH = '../../experiments/stopwords/cn_stopwords.txt'
DOC2VEC = '../model/doc2vec'
CLUSTERING_MODEL = '../model/clustering_model.joblib'
PCA_MODEL = '../model/pca_model.joblib'
RESULT = '../result/result.csv'

## doc2vec
# https://radimrehurek.com/gensim/models/doc2vec.html#usage-examples
VICTOR_SIZE=150
WINDOW=5
MIN_COUNT=3
NEGATIVE=5
EPOCHS=5
WORKERS=4

## kmeans
N_CLUSTERS = [2, 3, 4, 5, 6]
```

```python
df = pd.read_csv(DATA_PATH)
```

```python
# remove punctuations

punc_en = string.punctuation
punc_cn = punctuation
punc = punc_en + punc_cn

def remove_punc(x):
    for t in x:
        if t in punc:
            x=x.replace(t,'')
    return x

df['content_cut'] = df[CONTENT_FIELD].apply(lambda x: remove_punc(x))
```

```python
# tokenization
df['content_cut'] = df['content_cut'].apply(lambda x: jieba.lcut(x))
```

```python
# remove stopwords

stopwords_list = []
f = open(STOPWORDS_PATH)
lines = f.read()
stopwords_list = lines.split('\n')
f.close()

def remove_stopwords(x):
    x = [t for t in x if t not in stopwords_list]
    return x

df['content_cut'] = df['content_cut'].apply(lambda x: remove_stopwords(x))
```

```python
# # label encoding

# labels = df.class_label.value_counts().index.tolist()
# id2label = dict()
# label2id = dict()
# for i, lb in enumerate(labels):
#     id2label[i]=lb
#     label2id[lb]=i

# df['label_id'] = df[LABEL_FIELD].map(label2id)
```

```python
X_train = []
for i, word_list in enumerate(df.content_cut.tolist()):
    length = len(word_list)
    doc = gensim.models.doc2vec.TaggedDocument(word_list, [i])
    X_train.append(doc)

    
# Doc2Vec training
t1 = time.time()
model = gensim.models.doc2vec.Doc2Vec(
    X_train,
    vector_size=VICTOR_SIZE,
    window=WINDOW, 
    min_count=MIN_COUNT, 
    negative=NEGATIVE,
    epochs=EPOCHS,
    workers=WORKERS
)
t2 = time.time()
print("Time consumption:")
print(t2 - t1)
```

```python
model.save(DOC2VEC)
```

```python
model = gensim.models.doc2vec.Doc2Vec.load(DOC2VEC)  
```

```python
## Do doc2vec

t1 = time.time()
df['doc_vector'] = df['content_cut'].apply(lambda x:model.infer_vector(x) )
t2 = time.time()
print("Time consumption:")
print(t2 - t1)

doc_matrix = df['doc_vector'].tolist()
```

```python
# clustering, cross-validation

t1 = time.time()

k_s = []
scores = []

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(doc_matrix)
    y_pred = kmeans.predict(doc_matrix)
    score = silhouette_score(doc_matrix, y_pred)
    k_s.append(k)
    scores.append(score)
    
t2 = time.time()

print("Time consumption:")
print(t2 - t1)
```

```python
print("K, Scores:")
print(*zip(range(2, 7), scores))
```

```python
sns.lineplot(x=k_s, y=scores)
```

```python
## Clustering with PCA
pca = PCA(n_components=2)

doc_matrix_reduced = pca.fit_transform(doc_matrix)
pca.explained_variance_ratio_

dump(pca, PCA_MODEL)
```

```python
## Plot for anlysis with true label

print(__doc__)

X = doc_matrix_reduced
# y = df.label_id.tolist()

range_n_clusters = N_CLUSTERS

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

```python
# Save the best KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
dump(kmeans, CLUSTERING_MODEL)
```
